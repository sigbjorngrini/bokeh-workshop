# Bokeh plotting

First, we will go through how bokeh works to plot data

First we will import the packages and functions we need

```python
import pandas as pd
import numpy as np
from stats_tennis import player_stats

from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
```

Need to make sure it is outputed in the notebook

```python
output_notebook()
```

```python
matches = pd.read_csv('wta_matches_2018.csv', parse_dates=True)
stats = player_stats(matches)
```

Now we need to make it such that BokehJS will understand it. Then we make a ColumnDataSource.

```python
source = ColumnDataSource(stats)
```

## Making a figure

```python
fig = figure(plot_width=350, plot_height=350,
              tools='pan,wheel_zoom,box_select')
```

We show it in the notebook by using the show function

```python
show(fig)
```

We need to put in a line with some data

```python
line = fig.line(x='Rank', y='Wins', source=source)
```

```python
show(fig)
```

```python
fig2 = figure(plot_width=350, plot_height=350,
              tools='pan,wheel_zoom,box_select')
circles = fig2.circle(x='Rank', y='Wins', size=3, source=source)
```

```python
show(fig2)
```

```python
fig3 = figure(plot_width=350, plot_height=350,
              tools='pan,wheel_zoom,box_select')
circles2 = fig3.circle(x='Rank', y='Losses', size=3, source=source)
```

```python
from bokeh.layouts import row
```

```python
show(row(fig2, fig3))
```

```python
fig4 = figure(plot_width=350, plot_height=350,
              tools='pan,wheel_zoom,box_select,box_zoom,reset')
circles3 = fig4.circle(x='Rank', y='Wins', size=3, source=source,
                       selection_color="orange", alpha=0.6, nonselection_alpha=0.1, selection_alpha=0.4)
fig5 = figure(plot_width=350, plot_height=350,
              tools='pan,wheel_zoom,box_select,box_zoom,reset')
circles4 = fig5.circle(x='Rank', y='Losses', size=3, source=source,
                       selection_color="orange", alpha=0.6, nonselection_alpha=0.1, selection_alpha=0.4)
```

```python
show(row(fig4, fig5))
```

```python
from bokeh.plotting import output_file
```

```python
output_file("workshop.html")
show(row(fig4, fig5))
```
